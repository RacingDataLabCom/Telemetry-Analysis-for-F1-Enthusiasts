```python
year= 2023
gp = 'Saudi Arabia'
event = 'R'
session_race = fastf1.get_session(year, gp, event)
session_race.load()
```

```python
session_race.laps
```

```python
driver ='PER'
driver_laps = session_race.laps.pick_driver(driver)
driver_laps
```

```python
driver_laps.describe()
```

```python
Lap = 10
driver_specific_lap = driver_laps[driver_laps['LapNumber']==Lap].iloc[0]
driver_specific_lap
```

```python
Lap = 10
driver_specific_lap = driver_laps[(driver_laps['LapNumber'] == Lap)]
driver_specific_lap
```

```python
driver_car_data = driver_specific_lap.get_car_data()
driver_car_data
```

```python
from matplotlib import pyplot as plt
from fastf1 import plotting
time = driver_car_data['Time']
speed = driver_car_data['Speed']
plotting.setup_mpl()
fig, ax = plt.subplots(figsize=(15,5))
ax.plot(time, speed, label='Fast', color='red')
ax.set_xlabel('Time')
ax.set_ylabel('Speed[Km/h]')
ax.set_title('Driver: '+driver)
plt.legend()
plt.show()
```

```python
driver1 = 'PER'
driver2 = 'ALO'
driver1_laps = session_race.laps.pick_driver(driver1)
driver2_laps = session_race.laps.pick_driver(driver2)
fig, ax = plt.subplots(figsize=(15,5))
ax.plot(driver1_laps['LapNumber'], driver1_laps['LapTime'], color='red')
ax.plot(driver2_laps['LapNumber'], driver2_laps['LapTime'], color='cyan')
ax.set_title(driver1+" vs "+driver2)
ax.set_xlabel("Lap Number")
ax.set_ylabel("Lap Time")
plt.legend([driver1,driver2])
```
